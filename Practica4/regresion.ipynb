```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
import operator
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv('datos.csv')
datos.head()
```

```python
datos.shape
datos.info()
datos.describe()
```

```python
X=datos.drop('y',axis=1).values
y=datos.drop('x',axis=1).values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,shuffle=True, random_state=0)
```

```python
#Linear Regression with OLS
regresion_lineal = LinearRegression()
regresion_lineal.fit(X_train,y_train)
predic_ols=regresion_lineal.predict(X_test)

r2_score_OLS=r2_score(y_test,predic_ols)
mse_OLS= mean_squared_error(y_test,predic_ols)

print("R2_score: "+str(r2_score_OLS))
print("Mse: "+str(mse_OLS))

OLS_1 = ['Regresin Lineal con OLS', mse_OLS, r2_score_OLS]

#Grfica
plt.figure(figsize=(8, 6))
plt.scatter(X_test, y_test, color='blue', label='Datos reales')
plt.plot(X_test, predic_ols, color='red', linewidth=2, label='Regresin lineal')
plt.xlabel('Variable independiente (X)')
plt.ylabel('Variable dependiente (y)')
plt.title('Regresin lineal de 2ndo grado con OLS')
plt.legend()
plt.show()
```

```python
#Regresion lineal de grado 2 con OLS 
#1
poly_entre= PolynomialFeatures(2)
entre=poly_entre.fit_transform(X_train)

regre_lin_g2= LinearRegression()
regre_lin_g2.fit(entre,y_train)
predic_ols_2= regre_lin_g2.predict(poly_entre.transform(X_test))

r2_score_ols_2=r2_score(y_test,predic_ols_2)
mse_OLS_2=mean_squared_error(y_test,predic_ols_2)

print("R2_score: "+str(r2_score_ols_2))
print("Mse: "+str(mse_OLS_2))

OLS_2 = ['Regresin Lineal de grado 2 con OLS', mse_OLS_2, r2_score_ols_2]
#2
poly= PolynomialFeatures(2)
entre2=poly.fit_transform(X_train.reshape(-1,1))


regre_lin_g2= LinearRegression()
regre_lin_g2.fit(entre2,y_train)

x_test = np.linspace(X_test.min(), X_test.max(), 30).reshape(-1, 1)

predic_ols_2_2= regre_lin_g2.predict(poly_entre.transform(x_test))

r2_score_ols_2_2=r2_score(y_test,predic_ols_2)
mse_OLS_2_2=mean_squared_error(y_test,predic_ols_2)

#Grfica
plt.figure(figsize=(8, 6))
plt.scatter(X_test, y_test, color='blue', label='Datos reales')
plt.plot(x_test, predic_ols_2_2, color='red', linewidth=2, label='Regresin lineal')
plt.xlabel('Variable independiente (X)')
plt.ylabel('Variable dependiente (y)')
plt.title('Regresin lineal de 2ndo grado con OLS')
plt.legend()
plt.show()
```

```python
#Linear Regression of 3rd grade with OLS 

#1
poly_entre3=PolynomialFeatures(3)
entre3=poly_entre3.fit_transform(X_train)


regre_lin_g3=LinearRegression()
regre_lin_g3.fit(entre3,y_train)
predic_ols_3=regre_lin_g3.predict(poly_entre3.transform(X_test))

r2_score_ols_3=r2_score(y_test,predic_ols_3)
mse_OLS_3=mean_squared_error(y_test,predic_ols_3)

print("R2_score: "+str(r2_score_ols_3))
print("Mse: "+str(mse_OLS_3))

OLS_3 = ['Regresin Lineal de grado 3 con OLS', mse_OLS_3, r2_score_ols_3]

#2
poly= PolynomialFeatures(3)
entre3=poly.fit_transform(X_train.reshape(-1,1))


regre_lin_g3_2= LinearRegression()
regre_lin_g3_2.fit(entre3,y_train)

x_test = np.linspace(X_test.min(), X_test.max(), 30).reshape(-1, 1)

predic_ols_3_2= regre_lin_g3.predict(poly.transform(x_test))


#Grfica
plt.figure(figsize=(8, 6))
plt.scatter(X_test, y_test, color='blue', label='Datos reales')
plt.plot(x_test, predic_ols_3_2, color='red', linewidth=2, label='Regresin lineal')
plt.xlabel('Variable independiente (X)')
plt.ylabel('Variable dependiente (y)')
plt.title('Regresin lineal de 3er grado con OLS')
plt.legend()
plt.show()
```

```python
#Linear Regression with SGD

model= SGDRegressor(learning_rate='constant',eta0=0.0000001, max_iter=10000)
y_train2= np.ravel(y_train)
model.fit(X_train,y_train2)
predic_sgd= model.predict(X_test)

r2_score_sgd=r2_score(y_test,predic_sgd)
mse_sgd= mean_squared_error(y_test,predic_sgd)

print("R2_score: "+str(r2_score_sgd))
print("Mse: "+str(mse_sgd))

SGD_1 = ['Regresin Lineal con SGD', mse_sgd, r2_score_sgd]

#Grfica
plt.figure(figsize=(8, 6))
plt.scatter(X_test, y_test, color='blue', label='Datos reales')
plt.plot(X_test, predic_sgd, color='red', linewidth=2, label='Regresin lineal')
plt.xlabel('Variable independiente (X)')
plt.ylabel('Variable dependiente (y)')
plt.title('Regresin lineal con BGD')
plt.legend()
plt.show()
```

```python
#Linear Regression of 2nd grade with SGD

#1
poly_entre= PolynomialFeatures(2)
entre=poly_entre.fit_transform(X_train)
prueba=poly_entre.fit_transform(X_test)

model_2= SGDRegressor(learning_rate='constant',eta0=0.0000001, max_iter=10000)
y_train2= np.ravel(y_train)
model_2.fit(entre,y_train2)
predic_sgd_2= model_2.predict(prueba)

r2_score_sgd_2=r2_score(y_test,predic_sgd_2)
mse_sgd_2= mean_squared_error(y_test,predic_sgd_2)

print("R2_score: "+str(r2_score_sgd_2))
print("Mse: "+str(mse_sgd_2))

SGD_2 = ['Regresin Lineal de grado 2 con SGD', mse_sgd_2, r2_score_sgd_2]
#2
poly= PolynomialFeatures(2)
entre=poly.fit_transform(X_train.reshape(-1,1))

model_2_2= SGDRegressor(learning_rate='constant',eta0=0.0000001, max_iter=10000)
y_train3= np.ravel(y_train)
model_2_2.fit(entre,y_train3)

x_test = np.linspace(X_test.min(), X_test.max(), 30).reshape(-1, 1)
predic_sgd_2= model_2_2.predict(poly.transform(x_test))

r2_score_sgd_2_2=r2_score(y_test,predic_sgd_2)
mse_sgd_2_2= mean_squared_error(y_test,predic_sgd_2)



#Grfica
plt.figure(figsize=(8, 6))
plt.scatter(X_test, y_test, color='blue', label='Datos reales')
plt.plot(x_test, predic_sgd_2, color='red', linewidth=2, label='Regresin lineal')
plt.xlabel('Variable independiente (X)')
plt.ylabel('Variable dependiente (y)')
plt.title('Regresin lineal de 2ndo grado con SGD')
plt.legend()
plt.show()
```

```python
#Linear Regression of 3rd grade with SGD

poly_entre3=PolynomialFeatures(3)
entre3=poly_entre3.fit_transform(X_train)
prueba3=poly_entre3.fit_transform(X_test)

model_3=SGDRegressor(learning_rate='constant',eta0=0.0000001, max_iter=10000)
y_train=np.ravel(y_train)
model_3.fit(entre3,y_train)
predic_sgd_3=model_3.predict(prueba3)

r2_score_sgd_3=r2_score(y_test,predic_sgd_3)
mse_sgd_3= mean_squared_error(y_test,predic_sgd_3)

print("R2_score: "+str(r2_score_sgd_3))
print("Mse: "+str(mse_sgd_3))

SGD_3 = ['Regresin Lineal de grado 3 con SGD', mse_sgd_3, r2_score_sgd_3]

#2
poly3=PolynomialFeatures(3)
entre3=poly3.fit_transform(X_train.reshape(-1,1))


model_3_2=SGDRegressor(learning_rate='constant',eta0=0.0000001, max_iter=10000)
y_train3_2= np.ravel(y_train)
model_3_2.fit(entre3,y_train3_2)

x_test = np.linspace(X_test.min(), X_test.max(), 30).reshape(-1, 1)
predic_sgd_3_2=model_3_2.predict(poly3.transform(x_test))

r2_score_sgd_3_2=r2_score(y_test,predic_sgd_3_2)
mse_sgd_3_2= mean_squared_error(y_test,predic_sgd_3_2)

plt.figure(figsize=(8, 6))
plt.scatter(X_test, y_test, color='blue', label='Datos reales')
plt.plot(x_test, predic_sgd_3_2, color='red', linewidth=2, label='Regresin lineal')
plt.xlabel('Variable independiente (X)')
plt.ylabel('Variable dependiente (y)')
plt.title('Regresin lineal de 3er grado con SGD')
plt.legend()
plt.show()
```
